```python
import numpy as np
import matplotlib.pyplot as plt

def densidade_normal(x, mu_x, sigma_x):
    return (1 / np.sqrt(2 * np.pi * sigma_x**2)) * np.exp(-((x - mu_x)**2 / (2 * sigma_x**2)))

# Parâmetros
media = 0
desvio_padrao = 1
x_valores = np.linspace(-3, 3, 1000)
y_valores = densidade_normal(x_valores, media, desvio_padrao)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x_valores, y_valores, label="Densidade ")
plt.xlabel("x")
plt.ylabel("Densidade")
plt.title("Função de Densidade Gaussiana")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Função para plotar histogramas e boxplots
def plot_histograms_boxplots(data1, data2, label1, label2):
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 2, 1)
    plt.hist(data1, bins=30, edgecolor='k', alpha=0.7)
    plt.title(f'Histograma - {label1}')
    plt.subplot(2, 2, 2)
    plt.hist(data2, bins=30, edgecolor='k', alpha=0.7)
    plt.title(f'Histograma - {label2}')
    plt.subplot(2, 2, 3)
    sns.boxplot(data1)
    plt.title(f'Boxplot - {label1}')
    plt.subplot(2, 2, 4)
    sns.boxplot(data2)
    plt.title(f'Boxplot - {label2}')
    plt.tight_layout()
    plt.show()

# Função para calcular e exibir estatísticas descritivas
def calc_stats(data, dist_type):
    mean = np.mean(data)
    median = np.median(data)

    # Cálculo da moda
    mode_result = stats.mode(data)
    mode = mode_result.mode[0] if mode_result.mode.size > 0 else np.nan

    variance = np.var(data, ddof=1)
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)

    print(f'Estatísticas para {dist_type}:')
    print(f'Média: {mean}')
    print(f'Mediana: {median}')
    print(f'Moda: {mode}')
    print(f'Variância: {variance}')
    print(f'Assimetria: {skewness}')
    print(f'Curtose: {kurtosis}')
    print()

# Gerar e analisar dados com 1.000 amostras
n = 1000
uniform_data = np.random.uniform(-3, 6, n)
normal_data = np.random.normal(3, 2, n)

print('Análise com 1.000 amostras:')
plot_histograms_boxplots(uniform_data, normal_data, 'Distribuição Uniforme', 'Distribuição Normal')
calc_stats(uniform_data, 'Distribuição Uniforme')
calc_stats(normal_data, 'Distribuição Normal')

# Função para repetir a geração e análise dos dados
def repeat_analysis(n_samples):
    uniform_data = np.random.uniform(-3, 6, n_samples)
    normal_data = np.random.normal(3, 2, n_samples)

    print(f'Análise com {n_samples} amostras:')
    plot_histograms_boxplots(uniform_data, normal_data, 'Distribuição Uniforme', 'Distribuição Normal')
    calc_stats(uniform_data, 'Distribuição Uniforme')
    calc_stats(normal_data, 'Distribuição Normal')

# Repetir a análise com 10.000 e 100.000 amostras
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Definindo o número de amostras
n = 10000

# Gerar dados
uniform_data = np.random.uniform(-3, 6, n)
normal_data = np.random.normal(3, 2, n)

# Gerar a variável xSum
xSum = 2 * normal_data + uniform_data / 3

# Função para calcular e exibir estatísticas descritivas
def calc_stats(data, dist_type):
    mean = np.mean(data)
    median = np.median(data)

    # Cálculo da moda
    mode_result = stats.mode(data, keepdims=True)
    mode = mode_result.mode[0] if mode_result.mode.size > 0 else np.nan

    variance = np.var(data, ddof=1)
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)

    print(f'Estatísticas para {dist_type}:')
    print(f'Média: {mean}')
    print(f'Mediana: {median}')
    print(f'Moda: {mode}')
    print(f'Variância: {variance}')
    print(f'Assimetria: {skewness}')
    print(f'Curtose: {kurtosis}')
    print()

# Estatísticas para a variável xSum
print('Análise para xSum com 10.000 amostras:')
calc_stats(xSum, 'xSum')

# Função para plotar gráficos de dispersão
def plot_scatter_and_correlation(x1, x2, label1, label2):
    plt.figure(figsize=(12, 6))

    # Gráfico de dispersão XN vs xSum
    plt.subplot(1, 2, 1)
    plt.scatter(x1, x2, alpha=0.5)
    plt.xlabel(label1)
    plt.ylabel(label2)
    plt.title(f'Dispersão entre {label1} e {label2}')

    # Gráfico de dispersão XN vs xU
    plt.subplot(1, 2, 2)
    plt.scatter(normal_data, uniform_data, alpha=0.5)
    plt.xlabel('XN')
    plt.ylabel('xU')
    plt.title('Dispersão entre XN e xU')

    plt.tight_layout()
    plt.show()

# Gráficos de dispersão
plot_scatter_and_correlation(normal_data, xSum, 'XN', 'xSum')

# Função para calcular e exibir coeficientes de correlação
def calc_correlation(x1, x2, label1, label2):
    correlation = np.corrcoef(x1, x2)[0, 1]
    print(f'Coeficiente de correlação entre {label1} e {label2}: {correlation:.4f}')

# Cálculo dos coeficientes de correlação
calc_correlation(normal_data, xSum, 'XN', 'xSum')
calc_correlation(normal_data, uniform_data, 'XN', 'xU')
```
